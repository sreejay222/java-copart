# Linear Regression Begins
### Our Dataset doesn't have predicting column hence we need to create the predicting variable
## Example 
### column Grad_unemployment_rate is given hence we can fetch Grad_employment_rate from 1.
### for instance Grad_unemployment_rate = 0.08754
### hence Grad_employment_rate = 1-0.08754 = 0.91246
### Declare the variable name as Grad_employment_rate

```python
### Creating column for Graduate employment rate
df['Grad_employment_rate'] = 1-df['Grad_unemployment_rate']
```

```python
try:
    def verify_answer():
        
        #enter your verification code here.
        #return true if your checks pass
        #eg: 
        if round(df['Grad_employment_rate'][0],3) == 0.912: 
            return True
        else:
            return False
        

    ref_assert_var = verify_answer()
except Exception as e:
    print('Your assertion block throws error: ' + str(e))
else:
    if ref_assert_var:
        print('continue')
    else:
        print('The answer did not pass the test.')
```

```python
### picking up 4 independent variable to check from above correlation which are not highly co-related
X = df[['Grad_total','Grad_median','Grad_share','Grad_premium','Nongrad_unemployment_rate']]
## Assinging Grad_employment_rate as Y since it is the predicting variable
y = df[['Grad_employment_rate']]
```

```python
## Pairplot on independent variable to see the distribution of the data and whether they are normally distributed or not.
sns.pairplot(X)
```

# Data Training and Validation

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
print('Coefficients: \n', lm.coef_)
```

```python
try:
    def verify_answer():
        
        #enter your verification code here.
        #return true if your checks pass
        #eg: 
        if X_train['Grad_total'][0] == 9173: 
            return True
        else:
            return False
        

    ref_assert_var = verify_answer()
except Exception as e:
    print('Your assertion block throws error: ' + str(e))
else:
    if ref_assert_var:
        print('continue')
    else:
        print('The answer did not pass the test.')
```

```python
# Forecasting Output
```
